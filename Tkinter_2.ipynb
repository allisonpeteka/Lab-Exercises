```python
#basic dropdown menu
from tkinter import *

root = Tk()
root.title("Dropdown practice")
root.geometry("400x400")

item_list = ["Yellow", "Blue", "Green", "Purple", "Violet", "Black"]

def chosen_option():
    color = value.get()
    result_label = Label(root, text=f"You've chosen {color}")
    root.configure(background=color)
    result_label.pack()
    
value = StringVar()
value.set("Select a Color")
    
dropdown = OptionMenu(root, value, *item_list)
choose_button = Button(root, text="Click to choose", command=chosen_option)

dropdown.pack()
choose_button.pack()

root.mainloop()
```

```python
from tkinter import *
import requests
from tkinter import font
import json
import random

root = Tk()
root.title("Dropdown of Authors")
root.geometry("400x400")

def pick_authors():
    r = requests.get("https://api.quotable.io/quotes?limit=100")
    data = r.json()["results"]
    authors = list(set([quote["author"] for quote in data]))
    return authors

def author_quote(author, text_box):
    global quote
    text_box.delete(0.0, END)
    r = requests.get(f"https://api.quotable.io/quotes?author={author}")
    data = r.json()
    if "results" in data:
        quote = data["results"][0]["content"]
        text_box.insert(END, quote)

def dropdown():
    global authors_dropdown
    authors = pick_authors()
    random.shuffle(authors)
    value.set(authors[0])
    authors_dropdown = OptionMenu(root, value, *authors)
    authors_dropdown.pack()

def r_button():
    authors_dropdown.pack_forget()
    dropdown()
    author_quote(value.get(), text_box)

value = StringVar()
dropdown()
text_box = Text(root, height=4, width=20, bg="light blue", font="Elephant", padx=40, pady=40)
get_button = Button(root, text="Select an Author", command=lambda: author_quote(value.get(), text_box))
reset_button = Button(root, text="Click to Reset", command=r_button)
heading = Label(root, text="Click below to get an inspirational quote!")

text_box.pack()
get_button.pack()
reset_button.pack()
heading.pack()

root.mainloop()
```

```python
from tkinter import *
from PIL import ImageTk, Image
import requests
import json

lat = "40.732600"
lon = "-73.445401"
url = (f"https://api.weather.gov/points/{lat},{lon}")

def forecast():
    r = requests.get(url)
    json_file = r.json()
    json_file["@context"]
    forecast = json_file["properties"]["forecast"]
    new_request = requests.get(forecast)
    date = new_request.json()["properties"]["periods"]
    forecast_gui(date)
    
def forecast_gui(forecast_data):
    text.delete(1.0, END)
    for periods in forecast_data:
        weather_string = f"({periods['name']}) , ({periods['temperature']}) , ({periods['shortForecast']})"
        text.insert(END, weather_string + "\n")

root = Tk()
root.geometry("500x600")

canvas = Canvas(root, bg="white", height=100, width=500) #the new widget added, I chose this widget so it can look like a weather symbol on the top like on the weather app
canvas.pack()

img = Image.open("sunshine.jpeg")
img = img.resize((500, 200), Image.ANTIALIAS)
photo = ImageTk.PhotoImage(img)
canvas.create_image(0, 0, image=photo, anchor=NW)

text = Text(root, font=("Elephant", 12))
text.pack()

button = Button(root, text="refresh", command=forecast)
button.pack(side="bottom")

forecast()
root.mainloop()
```
