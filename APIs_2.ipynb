```python
data_file = open("weather_data.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(data_file)
csv_writer.writerow(["name", "temperature", "detailedForecast"])

date = json["properties"]["periods"]

for i in date:
    name2 = i["name"]
    temp = i["temperature"]
    dtemp = i["detailedForecast"]
    csv_writer.writerow([name2, temp, dtemp])
        
data_file.close()
```

```python
#fix
import requests
import json
import csv
import pandas as pd

country_codes = ['CA', 'IT','AL', 'CN', 'NL', 'PL', 'SM', 'TR', 'US', 'RU']
url1 = f"https://date.nager.at/api/v3"

data_file = open("public_holidays.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(data_file)
csv_writer.writerow(["Country Code", "Total Holidays"])

for country_code in country_codes:
    url2 = f"{url1}/PublicHolidays/2023/{country_code}"
    answer = requests.get(url2)
    holidays = answer.json()
    total_holidays = len(holidays)
    
    csv_writer.writerow([country_code, total_holidays])
    
data_file.close()
```
